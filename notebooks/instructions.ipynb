### 4. Enjoy the Simulation:
4 Robots deliver load objects (black cylinders) from the prey to the nest collaboratively. They handover their load to form a chain with one head, one tail and two intermediates. The approach is completely decentralized and uses neither global communication nor centralized control. Communication is established via colored LEDs on each robot. The chain is dynamic thus if one robot fails the other robots overtake its role and form a working transport chain again. The Web App visualizes the currently exposed Messages, Roles, Teams and some other useful parameters of the robots.
